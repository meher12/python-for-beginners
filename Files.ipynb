```python
reviews = []
lbl = []
for line in open("data/sentiment-labelled-sentences/yelp_labelled.txt", "r"):
    line = line.strip()
    if line != "":
        records = line.split("\t")
        reviews.append(records[0])
        lbl.append(records[1])
    
```

```python
# reviews[:10]
# lbl[10]
```

```python
res_file = open("data/sentiment-labelled-sentences/res.txt", "w", encoding='utf8')
res_file.truncate()
```

```python
for lb in lbl:
    if int(lb) == 1:
        res_file.write("positive " + "\n")
    else:
        res_file.write("negative " + "\n")
res_file.close()
```

```python
res_file = open("data/sentiment-labelled-sentences/res.txt", "a", encoding='utf8')
```

```python
for rev in reviews:
    res_file.write(str(len(rev)) + "\n")
res_file.close()
```
